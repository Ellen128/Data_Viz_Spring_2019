```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

```python
disc_sum = pd.DataFrame({
    'Total Schools':[0],
    'Total Students':[0],
    'Total Budget':[0],
    'Average Math Score':[0],
    'Average Reading Score':[0],
    '%Passing Math':[0.0],
    '%Passing Reading':[0.0],
    '%Overall Passing Rate':[0.0]})
disc_sum
```

```python
disc_sum['Total Schools'] = len(school_data_complete['school_name'].unique())
#school_data_complete['school_name'].value_counts()
disc_sum['Total Students'] = school_data_complete['student_name'].count()
disc_sum['Total Students']
```

```python
disc_sum['Total Budget'] = school_data['budget'].sum()
disc_sum['Total Budget'] = disc_sum['Total Budget'].map("${:.2f}".format)
disc_sum['Total Budget']
```

```python
disc_sum['Average Math Score'] = school_data_complete['math_score'].sum()/disc_sum['Total Students']
disc_sum['Average Math Score'] = disc_sum['Average Math Score'].map("{:.2f}".format)
disc_sum['Average Math Score']
```

```python
disc_sum['Average Reading Score'] = school_data_complete['reading_score'].sum()/disc_sum['Total Students']
disc_sum['Average Reading Score'] = disc_sum['Average Reading Score'].map("{:.2f}".format)
disc_sum['Average Reading Score']
```

```python
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
len(passing_math)
```

```python
disc_sum
```

```python
disc_sum['%Passing Math'] = len(passing_math)/disc_sum['Total Students']*100
disc_sum['%Passing Math']
```

```python
passing_read = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
disc_sum['%Passing Reading'] = len(passing_read)/disc_sum['Total Students']*100
disc_sum['%Passing Reading']
```

```python
#passing_overall = school_data_complete.loc[(school_data_complete['math_score'] + school_data_complete['reading_score'])/2 >= 70]
#len(passing_overall)
```

```python
disc_sum['%Overall Passing Rate'] = (disc_sum['%Passing Math'] + disc_sum['%Passing Reading'])/2
disc_sum['%Overall Passing Rate']
```
