# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Define the dataframe
disc_sum = pd.DataFrame({
    'Total Schools':[0],
    'Total Students':[0],
    'Total Budget':[0],
    'Average Math Score':[0],
    'Average Reading Score':[0],
    '%Passing Math':[0.0],
    '%Passing Reading':[0.0],
    '%Overall Passing Rate':[0.0]})
disc_sum
```

```python
disc_sum['Total Schools'] = len(school_data_complete['school_name'].unique())

disc_sum['Total Students'] = school_data_complete['student_name'].count()
disc_sum['Total Students']
```

```python
disc_sum['Total Budget'] = school_data['budget'].sum()
disc_sum['Total Budget'] = disc_sum['Total Budget'].map("${:,.2f}".format)
disc_sum['Total Budget']
```

```python
disc_sum['Average Math Score'] = school_data_complete['math_score'].sum()/disc_sum['Total Students']
disc_sum['Average Math Score'] = disc_sum['Average Math Score'].map("{:.2f}".format)
disc_sum['Average Math Score']
```

```python
disc_sum['Average Reading Score'] = school_data_complete['reading_score'].sum()/disc_sum['Total Students']
disc_sum['Average Reading Score'] = disc_sum['Average Reading Score'].map("{:.2f}".format)
disc_sum['Average Reading Score']
```

```python
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
len(passing_math)
```

```python
disc_sum['%Passing Math'] = len(passing_math)/disc_sum['Total Students']*100
disc_sum['%Passing Math']
```

```python
passing_read = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
disc_sum['%Passing Reading'] = len(passing_read)/disc_sum['Total Students']*100
disc_sum['%Passing Reading']
disc_sum['Average Math Score']
disc_sum['Average Reading Score']
```

```python
disc_sum['%Overall Passing Rate'] = (disc_sum['Average Reading Score'].astype(float) + disc_sum['Average Math Score'].astype(float))/2
disc_sum['%Overall Passing Rate']
```

```python
#Display District summary
disc_sum
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#school_data_complete["overall_score"] = (school_data_complete["reading_score"] +school_data_complete["math_score"])/2
```

```python
#school_data_indexed = school_data_complete.set_index(['school_name'])
```

```python
#group by to separate the data into different schools
group_school = school_data_complete.groupby(['school_name'])
group_school.count().head()
```

```python
schoolsum_df = pd.DataFrame({
    'School Type': group_school['type'].first(),                          
    'Total Students':group_school['student_name'].count(),
    'Total School Budget':(group_school['budget'].first()).map("${:,.2f}".format),
    'Per student Budget':group_school['budget'].first(),
    'Average Math Score':group_school['math_score'].mean(),
    'Average Reading Score':group_school['reading_score'].mean()})

schoolsum_df.head()
```

```python
##find the students with math score or reading score >= 70
math_pass = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]
read_pass = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
overall_pass = school_data_complete.loc[(school_data_complete["math_score"] >= 70) &
                                        (school_data_complete["reading_score"] >= 70)]
```

```python

schoolsum_df['%Passing Math'] = math_pass.groupby(['school_name'])['student_name'].count()
schoolsum_df['%Passing Reading']  = read_pass.groupby(['school_name'])['student_name'].count()
schoolsum_df['%Overall Passing Rate']  = overall_pass.groupby(['school_name'])['student_name'].count()

schoolsum_df.head()
```

```python
schoolsum_df['Per student Budget'] = (schoolsum_df['Per student Budget'] / schoolsum_df['Total Students']).map("${:,.2f}".format)
schoolsum_df['%Passing Math'] = schoolsum_df['%Passing Math'] / schoolsum_df['Total Students']*100
schoolsum_df['%Passing Reading']  = schoolsum_df['%Passing Reading'] / schoolsum_df['Total Students']*100
schoolsum_df['%Overall Passing Rate']  = (schoolsum_df['%Passing Math'] + schoolsum_df['%Passing Reading']) / 2

schoolsum_df.head()
```

```python
schoolsum_sorted_df = schoolsum_df.sort_values(["%Overall Passing Rate"], ascending=False)
schoolsum_sorted_df.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school

```python
group_grade = school_data_complete.groupby(['school_name','grade'])
grade_df = group_grade.count()
grade_df.head()

group_grade_df0 = school_data_complete.groupby(['school_name']).sum()
group_grade_df0.head()
```

```python
math0 = pd.DataFrame({
    'math_score':group_grade['math_score'].sum(),
    'students':group_grade['Student ID'].count()})
math0 = math0.reset_index()
math0.head()
```

```python
group_grade_df0['9th'] = list(math0.loc[np.arange(3, 60, 4),'math_score'])
group_grade_df0['10th'] = list(math0.loc[np.arange(0, 60, 4),'math_score'])
group_grade_df0['11th'] = list(math0.loc[np.arange(1, 60, 4),'math_score'])
group_grade_df0['12th'] = list(math0.loc[np.arange(2, 60, 4),'math_score'])

group_grade_df0['s9th'] = list(math0.loc[np.arange(3, 60, 4),'students'])
group_grade_df0['s10th'] = list(math0.loc[np.arange(0, 60, 4),'students'])
group_grade_df0['s11th'] = list(math0.loc[np.arange(1, 60, 4),'students'])
group_grade_df0['s12th'] = list(math0.loc[np.arange(2, 60, 4),'students'])

group_grade_df0.head()
```

```python
group_grade_df0['9th'] = group_grade_df0['9th'] / group_grade_df0['s9th']
group_grade_df0['10th'] = group_grade_df0['10th'] / group_grade_df0['s10th']
group_grade_df0['11th'] = group_grade_df0['11th'] / group_grade_df0['s11th']
group_grade_df0['12th'] = group_grade_df0['12th'] / group_grade_df0['s12th']
group_grade_df0.head()
math_grade_df = group_grade_df0.loc[:,['9th','10th','11th','12th']]
math_grade_df 
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
spending_df = schoolsum_df
spending_df.reset_index(drop=True,inplace=True)
spending_df.head()
```

```python
spending_df['Spending Ranges (Per Student)'] = pd.cut((spending_df["Per student Budget"]).replace(
    '\$', '', regex=True).astype('float'), spending_bins, labels=group_names,include_lowest=True)
spending_df.head()
```

```python
groupspending = spending_df.groupby('Spending Ranges (Per Student)')

groupspending_df = pd.DataFrame(groupspending['Average Math Score'].mean())#, groupspending['Average Reading Score'].mean()})
#groupspending_df = groupspending_df.drop(groupspending_df.columns[[0,1,2,3]], axis=1) #del groupspending_df[['School Type']#,'Total Students','Total School Budget','Per student Budget']]
groupspending_df.head()
groupspending_df['Average Reading Score'] = groupspending['Average Reading Score'].mean()
groupspending_df['%Passing Math'] =groupspending['%Passing Math'].mean()
groupspending_df['%Passing Reading'] =groupspending['%Passing Reading'].mean()
groupspending_df['%Overall Passing Rate'] =groupspending['%Overall Passing Rate'].mean()
groupspending_df
```
